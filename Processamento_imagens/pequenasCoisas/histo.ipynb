```python
import matplotlib.pyplot as plt

def plot_histogram(pixel_counts):
    # Quantidade de tons de cinza
    num_tons = len(pixel_counts)

    # Cria uma lista de índices para os tons de cinza
    tons_de_cinza = list(range(0, num_tons))

    # Cria o gráfico de barras
    plt.bar(tons_de_cinza, pixel_counts, color='blue')

    # Adiciona rótulos aos eixos
    plt.xlabel('Tons de Cinza')
    plt.ylabel('Quantidade de Pixels')
    plt.title('Histograma de Tons de Cinza')

    # Exibe o gráfico
    plt.show()

# Exemplo de lista de contagens de pixels
# Ajuste esta lista de acordo com seus dados
lista_de_pixeis = [8, 4, 3, 0, 0, 0, 4, 0]

# Chama a função para plotar o histograma
plot_histogram(lista_de_pixeis)
```
