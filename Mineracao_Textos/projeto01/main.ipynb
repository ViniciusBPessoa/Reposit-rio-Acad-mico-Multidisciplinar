```python
from sklearn.feature_extraction.text import CountVectorizer

# Exemplo de documentos
documents = [
    'O cachorro correu no parque.',
    'O gato dormiu na cama.',
    'O pássaro voou para longe.'
]

# Crie um objeto CountVectorizer
vectorizer = CountVectorizer()

# Ajuste o vetorizador ao texto e transforme os documentos em uma matriz de termos de documentos (bag-of-words)
X = vectorizer.fit_transform(documents)

# Obtenha o vocabulário
vocab = vectorizer.get_feature_names_out()

# Converta a matriz para uma representação de matriz densa
X_dense = X.toarray()

print("Vocabulário:", vocab)
print("Matriz de termos de documentos (bag-of-words):\n", X_dense)
```
